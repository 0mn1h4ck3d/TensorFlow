```python
!pip install pytrends
import pandas as pd
from pytrends.request import TrendReq

# Set up pytrends
pytrend = TrendReq(hl='pl-PL', tz=360)

# Provide your search terms
kw_list = ['matura', 'angielski']

# Set the timeframe to analyze today
timeframe = 'now 1-d'

# Build the payload
pytrend.build_payload(kw_list=kw_list, timeframe=timeframe, geo='PL')

# Get related queries
related_queries = pytrend.related_queries()

# Get the top related queries
top_queries = related_queries[kw_list[0]]['top']

# Get the rising related queries
rising_queries = related_queries[kw_list[0]]['rising']

# Extract the query phrases and their search percentages for top queries
top_query_phrases = top_queries['query']
top_search_percentages = top_queries['value']

# Extract the query phrases and their search percentages for rising queries
rising_query_phrases = rising_queries['query']
rising_search_percentages = rising_queries['value']

# Create a DataFrame to store the results
results_df = pd.DataFrame({'Top Wyszukiwania': top_query_phrases, 'Procent Wyszukiwan (Top)': top_search_percentages,
                           'Rosnace Wyszukiwania': rising_query_phrases, 'Procent Wyszukiwan (Rosnace)': rising_search_percentages})

# Save the results as a CSV file
results_df.to_csv('wyniki.csv', index=False)

# Display the results table
print(results_df)
```
